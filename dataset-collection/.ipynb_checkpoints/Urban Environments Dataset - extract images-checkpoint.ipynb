# Extract sample images

```python
locationFiles = glob.glob(outPath + "locations_*.csv")
locationFiles = {os.path.basename(x).split("_")[1].split(".")[0]:x for x in locationFiles}
locationFiles
```

```python
city = "boston"

NUM_TRIES = 5
```

```python
locationFile = locationFiles[city]

locations = pd.read_csv(locationFile).drop("Unnamed: 0", axis=1)
print "Extracting %d samples for %s."%(len(locations), city)

locations.head()
```

```python
locations.index.nunique()
```

```python
progressBar = FloatProgress(min=0, max=len(locations))
# display progress
display(progressBar)
```

```python
i = -1
# i += -1
while i < len(locations)-1:
    i += 1
    clear_output(wait=True)
    print "%d/%d"%(i, len(locations))
    progressBar.value += 1
    
    sampleId, (imgLon, imgLat, area, _) = locations.index[i], tuple(locations.iloc[i].values)
    print "Extracting images around %s (%2.4f, %2.4f)"%(area, imgLat, imgLon)
    
#     # get Google Maps image (with error handling since Google might return error)
#     gmFileName = "%s/samples/%s/%d/googlemaps_%2.5f_%2.5f.jpg"%(outPath, city, sampleId, imgLat, imgLon) 
#     req = gmClient.construct_static_url(center="%2.5f,%2.5f"%(imgLat, imgLon), maptype="satellite", zoom=17)
#     print req
#     numTries = 0
#     while numTries < NUM_TRIES:
#         numTries += 1
#         try:
#             imgGM = gmClient.get_static_map_image(req, filename=gmFileName)
#             break
#         except:
#             print "Error! Trying again (%d/%d) in 5 sec"%(numTries, NUM_TRIES)
#             time.sleep(5)                
#     time.sleep(1.0/40.0)
#     # hack to test for the gray error image that Google returns when the API daily quota has been exceeded
#     if imgGM is None or np.abs(imgGM.mean() - 223.53)<0.01:
#         print "Exhausted API calls on the Google key %s. Image id = %d, sample id = %d"%(gmClient._key, i, sampleId)
#         break

    # get raster info at 3 scales
    sarBaseName = "%s/%s/%d/SAR_"%(outPath, city, sampleId) 
    vals = {}
    for kmScale in [1,5,10]:
        try:
            imgSAR = sarImgClient.get_image_at_location((imgLat, imgLon), w=kmScale, \
                                                        dumpPath=sarBaseName)[0]
            imgNL  = nightImgClient.get_image_at_location((imgLat, imgLon), w=kmScale)[0]
            imgPop = popImgClient.get_image_at_location((imgLat, imgLon), w=kmScale)[0]
            nlPixVal = imgNL[imgNL>=0].mean()
            popPixVal = imgPop[imgPop>=0].mean()  
            vals[kmScale] = (nlPixVal, popPixVal)
        except:
            print "Error!"
            continue

    # save current pixel to file
    savepath = "%s/%s/%d/"%(outPath, city, sampleId)
    if not os.path.exists(savepath):
        continue
    df = pd.DataFrame(vals,index=["avg nightlights", "avg population"]).T
    df.insert(0,"sample ID", sampleId)
    df.insert(1,"city", city)
    df.insert(2,"area", area)
    df.insert(3,"img lat", imgLat)
    df.insert(4,"img lon", imgLon)
    df.insert(5,"scale km", df.index)
    df.reset_index(inplace=True)
    df.drop("index",1,inplace=True)
    df.to_csv("%s/pixel_data.csv"%savepath, index=False)
    
    # display samples every now and then
    if i % 100 == 0:
        plt.figure(figsize=(16,10))
        plt.subplot(1,3,1)
#         # google maps
#         plt.imshow(imgGM)
#         plt.title("Google Maps Image %d"%i)
        # SAR
        plt.imshow(imgSAR)
        plt.title("SAR Image %d"%i)
        # night lights
        plt.subplot(1,3,2)
        plt.imshow(imgNL)
        plt.title("Night Lights Image %d"%i)
        # population
        plt.subplot(1,3,3)
        plt.imshow(imgPop)
        plt.title("Population Image %d"%i)
        plt.show()
        time.sleep(5)
```
