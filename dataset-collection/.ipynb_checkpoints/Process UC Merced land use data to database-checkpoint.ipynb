```python
datapath = "/home/adalbert/data/UCMerced_LandUse/"

files = glob.glob(datapath + "Images/*/*.tif")
files = [(f,f.split("/")[-2]) for f in files]
df = pd.DataFrame(files, columns=["filename", "class"])

df.head()
```

```python
train_file = "%s/train.csv"%(datapath)
test_file = "%s/test.csv"%(datapath) 

if not os.path.exists(train_file) or not os.path.exists(test_file):
    TRAIN_FRAC = 0.95
    RAND_SEED = 0
    # split to train/test
    df_train = df.sample(n=int(TRAIN_FRAC*len(df)), random_state=RAND_SEED)
    df_test = df.ix[~df.index.isin(df_train.index)]
    # save to file
    df_train.to_csv(train_file)
    df_test.to_csv(test_file)
else:
    df_train = pd.read_csv(train_file)
    df_test = pd.read_csv(test_file)
    
print df_train.shape, df_test.shape
```

```python
train_db_path = "%s/leveldb-96-norm/train"%(datapath)
test_db_path = "%s/leveldb-96-norm/test"%(datapath)

if not os.path.exists(train_db_path):
    os.makedirs(train_db_path)
    sourceList = map(lambda x: (x,), df_train['filename'].values)
    cu.save_images_to_leveldb(sourceList, train_db_path, imgSize=(96,96))

if not os.path.exists(test_db_path):
    os.makedirs(test_db_path)
    sourceList = map(lambda x: (x,), df_test['filename'].values)
    cu.save_images_to_leveldb(sourceList, test_db_path, imgSize=(96,96))
```

```python
X,Y = cu.read_from_leveldb(train_db_path, start=10, stop=20, float_data=True)
```
