# Plot samples

```python
import matplotlib.gridspec as gridspec
from scipy.misc import imread
import skimage

def plot_sample(cols):
    # read in data
    nCols = len(cols) 
    gs1 = gridspec.GridSpec(1, nCols)
    gs1.update(wspace=0.05, hspace=0.05) # set the spacing between axes. 
    axes = np.array([plt.subplot(gs1[j]) for j in range(nCols)])
    
    for ax in axes.flatten():
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.grid(False)

    for ax, c in zip(axes, cols.keys()):
        ax.set_title(c, fontsize=16)
    
    # plot grid of images
    for i,c in enumerate(cols.values()):
        img = imread(c)
        axes[i].imshow(img)
            
    plt.show()
```

```python
sampleID = 9

img_src = dict(zip(df[df.index==sampleID]['scale km'], \
              df[df.index==sampleID]['filename']))
plot_sample(img_src)
```
